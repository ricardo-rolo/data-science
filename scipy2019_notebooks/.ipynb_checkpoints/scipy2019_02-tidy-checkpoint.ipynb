# 1. Columns containing values, not variables

```python
import pandas as pd
```

```python
pew = pd.read_csv('../data/pew.csv')
```

```python
pew.head()
```

```python
pew.melt(id_vars='religion')
```

```python
pew_tidy = pew.melt(id_vars='religion',
        var_name='income',
        value_name='count')
```

```python
pew_tidy.head()
```

```python
billboard = pd.read_csv('../data/billboard.csv')
```

```python
billboard.head()
```

```python
(billboard
 .melt(id_vars=['year','artist','track','time','date.entered'],
       var_name='week',
       value_name='rank')
 .groupby('artist')['rank']
 .mean()
)
```

# 2. Multiple variables stored in one column

```python
ebola = pd.read_csv('../data/country_timeseries.csv')
```

```python
ebola.head()
```

```python
ebola_long = (ebola
              .melt(id_vars=['Date','Day'],
                   var_name='cd_country',
                   value_name='count')
             )
```

```python
ebola_long.head()
```

```python
ebola_split = ebola_long['cd_country'].str.split('_', expand=True)
```

```python
ebola_long[['status','country']] = ebola_split
```

```python
ebola_long
```

```python
(ebola_long
 .groupby(['country','status'])['count']
 .mean()
 .reset_index()
)
```

# 3. Variables are stored in both rows and columns

```python
weather = pd.read_csv('../data/weather.csv')
```

```python
weather.head()
```

```python
weather_long = weather.melt(id_vars=['id','year','month','element'],
             var_name='day',
             value_name='temp')
```

```python
(weather_long
 .pivot_table(index=['id','year','month','day'],
              columns='element',
              values='temp')
 .reset_index())
```
